```python
#Create a copy 
df1 = df.copy()
```

```python
#Make sure there is no space in the column names
df1.columns = df1.columns.str.strip()
```

```python
# We only need the course id(without CRN and term) (example: ALY6000)
df1['course_id'] = df1['course_id'].str.slice(0, -13)
```

```python
#Making sure the change is reflected in the df1 dataframe
df1.drop(['course_name'], axis=1, inplace=True)
```

```python
df1.sample(n=2)
```

```python
df2 = df1['los'].str.split(('( ?\| ?)+'), expand=True) 
df2.columns = ['los{}'.format(x+1) for x in df2.columns] 
```

```python
df2.sample(n=55)
```

```python
#https://www.w3resource.com/pandas/series/series-reset_index.php
#https://stackoverflow.com/questions/32801806/pandas-concat-ignore-index-doesnt-work
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)

df3 = pd.concat( [df1, df2], axis=1) 
```

```python
df3.sample(n=5)
```

```python
df3["title"] = df3["title"].str.lower()
```

```python
df3.title.str.split(expand=True).stack().value_counts()
```

```python
df3.drop(['title'], axis=1, inplace=True)
```

```python
#https://stackoverflow.com/questions/13445241/replacing-blank-values-white-space-with-nan-in-pandas
df3.replace(r'^\s*$', np.nan, regex=True, inplace = True)
```

```python
#https://stackoverflow.com/questions/23743460/replace-none-with-nan-in-pandas-dataframe
df3.dropna(axis=1, how='all', inplace=True)
```
